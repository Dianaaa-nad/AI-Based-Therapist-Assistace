```python
import numpy as np
import pandas as pd

data = pd.read_csv('Suicide_Detection.csv')

data
```

**Remove stop words**


```python
import text_hammer as th
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize 
nltk.download('stopwords')
nltk.download('punkt')
lists = stopwords.words('english')
```

```python
stopwords_list = set(stopwords.words('english'))

def remove_stopwords(text):
  tokens = word_tokenize(text)
  tokens_without_stopwords = [word for word in tokens if word.lower() not in stopwords_list]
  text_without_stopwords = " ".join(tokens_without_stopwords)
  return text_without_stopwords

data['text'] = data['text'].apply(remove_stopwords)
data.head()
```

**Preprocessing**


```python
import re

def remove_emails(text):
    # Manual implementation to remove emails
    return re.sub(r'\S+@\S+', '', text)

def remove_html_tags(text):
    # Manual implementation to remove HTML tags
    return re.sub(r'<.*?>', '', text)

def remove_special_chars(text):
    # Manual implementation to remove special characters
    return re.sub(r'[^a-zA-Z0-9\s]', '', text)

def remove_accented_chars(text):
    # Manual implementation to remove accented characters
    # You can use unidecode or other methods for a more comprehensive solution
    return re.sub(r'[^\x00-\x7F]+', '', text)

def text_preprocessing(df, col_name):
    column = col_name
    df[column] = df[column].apply(lambda x: str(x).lower())
    df[column] = df[column].apply(lambda x: remove_emails(x))
    df[column] = df[column].apply(lambda x: remove_html_tags(x))
    df[column] = df[column].apply(lambda x: remove_special_chars(x))
    df[column] = df[column].apply(lambda x: remove_accented_chars(x))
    return df

data = text_preprocessing(data, 'text')
data.head()
```

**Tokenizer**


```python
data
```

```python
data2 = data.copy()

data2
```

```python
def custom_tokenizer(text):
    tokens = word_tokenize(text)
    return tokens

def tokenize_columns(df, text_column):
    df[text_column] = df[text_column].apply(lambda x: custom_tokenizer(str(x).lower()))
    return df

tokenize_data = tokenize_columns(data2, 'text')
print(tokenize_data[['text']])

tokenize_data.head()
```

```python
tokenize_data2 = tokenize_data.copy()

tokenize_data2
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import ComplementNB

# Join the tokenized words into sentences (documents)
documents = tokenize_data2['text'].apply(lambda x: ' '.join(x))

documents
```

```python

# Use CountVectorizer to convert the text data into a bag-of-words representation
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(documents)

# Map 'class' column to numeric values
y = data2['class'].map({'non-suicide': 0, 'suicide': 1})
```

```python

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=19)

# Create and train the Complement Naive Bayes model
cnb = ComplementNB()
cnb.fit(X_train, y_train)

# Evaluate the model
train_accuracy = cnb.score(X_train, y_train)
test_accuracy = cnb.score(X_test, y_test)

print("Train accuracy: {:.2f}%".format(train_accuracy * 100))
print("Test accuracy: {:.2f}%".format(test_accuracy * 100))
```

```python
import joblib
# Save the trained Complement Naive Bayes model
model_filename = 'complement_naive_bayes_model.joblib'
joblib.dump(cnb, model_filename)
print(f'Model saved to {model_filename}')

# Save the CountVectorizer
vectorizer_filename = 'count_vectorizer.joblib'
joblib.dump(vectorizer, vectorizer_filename)
print(f'CountVectorizer saved to {vectorizer_filename}')
```
